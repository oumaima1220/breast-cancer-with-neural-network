```python
# installation des packages
from sklearn import datasets
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# télécharger la base de donnée
df=datasets.load_breast_cancer()
```

```python
# découvrir le base de donnée
print(df.DESCR)
```

```python
df.target
```

```python
df.data
```

```python
# préparer les données
x=df.data
y=df.target
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, x_test.shape)
```

```python
# construction de réseau de neurone
# création du réseau de neurone
model=tf.keras.models.Sequential()
```

```python
# ajouter une couche de type dense, nombre des noeuds est 15(la taille de la couche caché intermédiare), fonction d'activation 
#pour couche caché esr RELU, pour sortie est SEGMOID, en spcifie les poids, la taille de la couche d'entrée est 30
model.add(tf.keras.layers.Dense(units=15,activation='relu',kernel_initializer='uniform',input_dim=30))
```

```python
# ajout de la couche de sortie,dimension est une seul unité, fonction d'activation pour couche de sortie est SEGMOID
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid',kernel_initializer='uniform'))
```

```python
# la fonction de cout(perte) loss, pour comparer la sortie avec la sortie réelle et faire mise à jour des poids, 
#precision du model: metrics
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['binary_accuracy'])
```

```python
# voir mon premier réseau de neuronne
model.summary()
```

```python
# entrainement de réseau de neuronne,lors de la mise à jour des poids faire passer toute le base de donné( batch_size=None), 
#faire passer 10 échontillons( batch_siez=10).
# epochs=combien de fois je vais passer la base de données pour entrainer mon model, donc pour chaque epochs je vais calculer 
#la la fct de cout loss et la metrics.
model.fit(x_train,y_train,batch_size=10,epochs=100)
# le cout va diminuer et la précision va augmenter
```

```python
#tester et evaluer le réseau de neuronne
test_loss, test_accuracy=model.evaluate(x_test,y_test)
print(f'Test loss {test_loss} et Test accuracy {test_accuracy}')
# model bien précis
```
